## EDA ON ZOMATO DATASET

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Zomato.csv', encoding = 'latin-1')
```

```python
df.isnull().sum()
```

```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df, df_country, on = 'Country Code', how = 'left')
```

```python
final_df.head()
```

```python
final_df.dtypes
```

```python
country_name = final_df.Country.value_counts().index
```

```python
country_values = final_df.Country.value_counts().values
country_values
```

```python
plot = plt.pie(country_values[0:3], labels = country_name[0:3], autopct = '%1.2f%%' )
```

Observation : The maximum amount of transactions or order on Zomato is from India followed by USA and UK.

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
ratings
```

```python
rating = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns= {0: 'Rating Count'})
```

```python
rating
```

```python
rating.head()
```

```python
plt.rcParams['figure.figsize'] = (12,6)
sns.barplot(x = 'Aggregate rating', y = 'Rating Count', data = rating)
```

```python
sns.barplot(x = 'Aggregate rating', y = 'Rating Count',hue = 'Rating color', data = rating, palette =['black', 'red', 'orange', 'yellow', 'green', 'violet'])
```

## Observation 
1. 0 rating or not rated count is very high.
2. Maximum amount of rating falls between 2.7 to 3.9
